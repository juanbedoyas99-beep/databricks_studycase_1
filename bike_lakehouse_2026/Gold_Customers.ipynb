```python
import pyspark.sql.functions as F
from pyspark.sql.window import Window
```

```python
query = """
select 
    ROW_NUMBER() OVER (ORDER BY ci.customer_id) AS customer_key,
    ci.customer_id,
    ci.first_name,
    ci.last_name
FROM bike_store.silver_data.crm_customer ci
"""

df = spark.sql(query)
```

```python
# write in the gold layer
df.write.mode("overwrite").saveAsTable("bike_store.gold_data.customer")
```
