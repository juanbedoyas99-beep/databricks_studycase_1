Reading from csv file 
Data loading into the bonze layer, no transforation are made here

```python
df = (
    spark.read.option("header", "true")
    .option("inferSchema","true")
    .csv("/Volumes/bike_store/bonze_data/source_system/source_crm/cust_info.csv")
)
df.display()
```

Write it to Bronze layer

```python
df.write.mode("overwrite").saveAsTable("bike_store.bonze_data.crm_cust_info")
```
